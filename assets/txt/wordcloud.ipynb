```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
text = open(r'C:\Users\e0306210\Documents\GitHub\kebinwebsite\assets\txt\abstract2009.txt').read()
```

```python
alice_coloring = np.array(Image.open(r'C:\Users\e0306210\Documents\GitHub\kebinwebsite\images\usa.png'))
stopwords = set(STOPWORDS)
stopwords.add("said")
stopwords.add("firm")
```

```python
wc = WordCloud(scale=5,background_color="white", max_words=2000, mask=alice_coloring,
               stopwords=stopwords, max_font_size=60, random_state=20)
# generate word cloud
wc.generate(text)
```

```python
image_colors = ImageColorGenerator(alice_coloring)
```

```python
fig, axes = plt.subplots(1, 1)
axes.imshow(wc, interpolation="bilinear")
# recolor wordcloud and show
# we could also give color_func=image_colors directly in the constructor
#axes[1].imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
#axes[2].imshow(alice_coloring, cmap=plt.cm.gray, interpolation="bilinear")
#for ax in axes:
plt.axis("off")
```

```python
wc.to_file(r'C:\Users\e0306210\Documents\GitHub\kebinwebsite\images\result_usa.jpg')
```
